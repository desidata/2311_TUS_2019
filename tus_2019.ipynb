```python
import pandas as pd

# Assuming df is your DataFrame

# Combine 'hour' and 'minute' columns to create a 'time' column
df['time'] = df['time_from_hrs'].astype(str).str.zfill(2) + ':' + df['time_from_mins'].astype(str).str.zfill(2)

# Combine 'date' and 'time' columns to create 'datetime' column
df['datetime'] = pd.to_datetime(df['time'].astype(str) + ' ' + df['time'])

# Drop the original 'hour' and 'minute' columns if needed
df = df.drop(['time_from_hrs', 'time_from_mins'], axis=1)

# Shift datetime values by 4 hours to account for the day starting at 4 am
df['adjusted_datetime'] = df['datetime']#+ pd.DateOffset(hours=4)

# Adjust the date if the time is less than 4 am
# NSSO survey takes place between 4 am to 4 am (starts from (n-1)th day to nth day). 
# So we have to shoft the times that are post midnight by 1 day
df.loc[df['datetime'].dt.hour < 4, 'adjusted_datetime'] += pd.DateOffset(days=1)

# Convert the activities to string
df['act'] = df['act'].astype(str)

# Convert 3, 4, 6 activity to a single activity
# Combine 'party', 'drink', and 'dance' into 'social'
df['act'] = df['act'].replace(['3', '4', '5'], '12')

# Print the modified DataFrame
df
```

```python
# Filters dates that are less than 4 am (sanity check)
df[df['adjusted_datetime']>='2023-11-27 00:00:00']
```

```python
# filter cols
df = df[['PID', 'act', 'adjusted_datetime']]
df.rename(columns = {'PID':'sample_id', 'act':'state',
                              'adjusted_datetime':'datetime'}, inplace = True)


df
```

```python
# Mapping of numeric codes to words
state_mapping = {'1': 'Employment', 
                 '2': 'Production for self consumption',
                 '12': 'Unpaid service',
                 '6': 'Learning',
                 '7':'Socialize/Communication/Religious activity',
                 '8':'Eclectic*',
                 '9':'Self care'}

# Replace numeric codes with words
df['state'] = df['state'].map(state_mapping)
```

```python
df
```

```python
# Function to fill missing rows between recorded activities
def fill_missing_rows(group):
    start_time = group['datetime'].min()
    end_time = group['datetime'].max()
    
    # Generate a DatetimeIndex for every minute between start and end
    minutes_range = pd.date_range(start=start_time, end=end_time, freq='1min')
    
    # Create a DataFrame with the generated DatetimeIndex and fill 'activity' with the last known activity
    result = pd.DataFrame({'datetime': minutes_range, 'state': group['state'].iloc[-1]})
    
    return result

unique_sample_id = list(df['sample_id'].unique())
master_df=[]
init_date=[]
last_time=[]
for i in unique_sample_id:
    
    filt_df=df[df['sample_id']==i]
    filt_df=filt_df.sort_values(by='datetime')
    #print(list(filt_df['datetime'])[0])
    init_date.append(list(filt_df['datetime'])[0])
    last_time.append(list(filt_df['datetime'])[-1])
    # Apply the function to each person_id group
    # Find the first datetime in the DataFrame
    first_datetime = filt_df['datetime'].iloc[0]

    # Create a new row with the same data but with the time set to 00:00:00
    new_row = {'sample_id': filt_df['sample_id'].iloc[0], 'datetime': first_datetime.replace(hour=0, minute=0, second=0), 'state': filt_df['state'].iloc[0]}

    # Add the new row to the DataFrame
    #filt_df = pd.concat([pd.DataFrame([new_row]), filt_df], ignore_index=True)
    
    # now we fill the rows
    filled_rows = filt_df.groupby('sample_id').apply(fill_missing_rows)

    # Merge the original DataFrame with the filled rows
    df_filled = pd.merge(filled_rows, filt_df, on=['sample_id', 'datetime'], how='left')
    # Forward fill the 'activity' column to fill NaN values with the last known activity
    df_filled['state'] = df_filled['state_y'].fillna(method='ffill')

    # Drop unnecessary columns
    df_filled = df_filled[['sample_id', 'datetime', 'state']]
    
    
    master_df.append(df_filled)

    
master_df = pd.concat(master_df, ignore_index=True)
master_df
```

```python
# Import library.
from d3blocks import D3Blocks

# Initialize.
d3 = D3Blocks()

# Make the moving bubbles
d3.movingbubbles(master_df, 
                 datetime='datetime',
                 state='state',
                 sample_id='sample_id',
                 size=3,
                 center='Self care',
                 #color=color,
                 color_method='node',
                 timedelta='minutes',
                 speed={"slow": 1000, "medium": 100, "fast": 10},
                 #time_notes=time_notes,
                 fontsize= 15, 
                 damper = 0.5,
                 filepath=r"C:\Users\trina\OneDrive\Documents\desidata\movingbubbles1.html",
                 cmap='Set1',
                 standardize='statewise',
                 figsize = [780, 800],
                 #title = "How Indians live\n(NSSO TUS 2019)",
                 note = 'We randomly sampled 1000 people accross all states, genders, age and living environment(i.e. rural or urban) from NSSO TUS 2019.*The Eclectic category primarily includes cultural, leisure, meass-media and sports.'
                 )

```
